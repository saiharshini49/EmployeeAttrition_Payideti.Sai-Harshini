###TASK 4-MODEL BUILDING & COMPARISON
   In this task, the dataset is divided into training and testing sets. Three machine learning models are trained and compared to identify the model that predicts employee attrition most effectively

```python
#Split the Data(80% Training,20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.20,random_state=42,stratify=y)
print("Training Data:", X_train.shape)
print("Testing Data:", X_test.shape)
```

```python
### Observation
The dataset was split into 80% training data and 20% testing data.
Stratified sampling was used to preserve the proportion of employees who stayed and left in both datasets.
```

```python
#Train Logistic Regression
log_model = LogisticRegression(class_weight='balanced',random_state=42,max_iter=1000)
log_model.fit(X_train, y_train)
```

```python
#train Random Forest
rf_model = RandomForestClassifier(n_estimators=100,class_weight='balanced',random_state=42)
rf_model.fit(X_train,y_train)
```

```python
#train Gradiant Boosting
gb_model = GradientBoostingClassifier(
    random_state=42
)

gb_model.fit(X_train, y_train)
```

```python
#Make Predictions
log_pred = log_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
gb_pred = gb_model.predict(X_test)
```

```python
#Compare Model Accuracy
from sklearn.metrics import accuracy_score
comparison = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "Accuracy": [
        accuracy_score(y_test, log_pred),
        accuracy_score(y_test, rf_pred),
        accuracy_score(y_test, gb_pred)
    ]
})
comparison
```

###Task5-Model Evaluation

```python
# Task 5: Model Evaluation
The trained machine learning models are evaluated using different performance metrics to determine which model predicts employee attrition most accurately.
```

```python
#Classification Report
print("Logistic Regression")
print(classification_report(y_test, log_pred))
```

```python
#Random Forest
print("Random Forest")
print(classification_report(y_test, rf_pred))
```

```python
#Gradient Boosting
print("Gradient Boosting")
print(classification_report(y_test, gb_pred))
```

```python
##Step 2:ROC-AUC Score
log_auc = roc_auc_score(y_test,log_model.predict_proba(X_test)[:,1])
rf_auc = roc_auc_score(y_test,rf_model.predict_proba(X_test)[:,1])
gb_auc = roc_auc_score(y_test,gb_model.predict_proba(X_test)[:,1])
print("Logistic Regression:",round(log_auc,3))
print("Random Forest:",round(rf_auc,3))
print("Gradient Boosting:",round(gb_auc,3))
```

```python
#Comparison Table
comparison = pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "ROC-AUC":[
        log_auc,
        rf_auc,
        gb_auc
    ]
})
comparison
```

# Bonus: ROC Curve Comparison

The Receiver Operating Characteristic (ROC) Curve is used to compare the performance of all three classification models. The model with the highest Area Under the Curve (AUC) demonstrates better ability to distinguish between employees who are likely to leave and those who are likely to stay.

```python
# Plot ROC Curves for all models

plt.figure(figsize=(8,6))

# Logistic Regression
fpr_log, tpr_log, _ = roc_curve(y_test, log_model.predict_proba(X_test)[:,1])
plt.plot(fpr_log, tpr_log,
         label=f"Logistic Regression (AUC = {log_auc:.3f})")

# Random Forest
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_model.predict_proba(X_test)[:,1])
plt.plot(fpr_rf, tpr_rf,
         label=f"Random Forest (AUC = {rf_auc:.3f})")

# Gradient Boosting
fpr_gb, tpr_gb, _ = roc_curve(y_test, gb_model.predict_proba(X_test)[:,1])
plt.plot(fpr_gb, tpr_gb,
         label=f"Gradient Boosting (AUC = {gb_auc:.3f})")

# Random Classifier
plt.plot([0,1],[0,1],'k--',label="Random Guess")

plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.legend(loc="lower right")
plt.grid(True)

# Save chart
plt.savefig("charts/chart6_roc_curve.png", dpi=300, bbox_inches="tight")

plt.show()
```

### Observation

The ROC Curve compares the classification performance of Logistic Regression, Random Forest, and Gradient Boosting models. A model with a higher Area Under the Curve (AUC) has better predictive performance. Based on the ROC-AUC scores obtained in this project, **Logistic Regression** achieved the best performance and was selected as the final model for predicting employee attrition.

# Task 7 – HR Insights & Business Recommendations
#### 1. Which 3 factors most strongly predict that an employee will leave?

Based on the Logistic Regression model, the three strongest factors influencing employee attrition are:
- Job Satisfaction
- Monthly Income
- Overtime

Employees with lower job satisfaction, lower monthly income, and those who frequently work overtime are more likely to leave the company.

---

#### 2. Which department or job role should HR prioritize for retention efforts?

The Sales Department and job roles with the highest attrition rate should be given priority for retention initiatives. Employees in these roles experience higher turnover and require focused engagement from HR.

---

#### 3. Does salary alone explain attrition?

No. Although lower monthly income contributes to employee attrition, salary alone does not explain why employees leave. Factors such as overtime, job satisfaction, work-life balance, years at the company, and career growth opportunities also have a significant impact.

---

#### 4. Two HR Recommendations

1. Introduce regular employee engagement and career development programs, especially for employees with low job satisfaction and high overtime.

2. Identify employees in high-risk departments and conduct periodic one-on-one retention discussions to address concerns before they decide to leave.

---

#### 5. Limitation of the Model

This model is trained on historical employee data and may not capture personal or external reasons for leaving, such as family circumstances, health issues, or better job opportunities. Therefore, the model should be used as a decision-support tool rather than the sole basis for HR decisions.
